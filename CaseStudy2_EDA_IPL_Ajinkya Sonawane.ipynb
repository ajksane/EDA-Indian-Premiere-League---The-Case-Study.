# Handling Missing Values

```python
df_ipl['city'].fillna('Dubai Sports City',inplace=True)
```

```python
df_ipl.drop(columns={'umpire3','umpire1','umpire2','city','id'},inplace=True)
```

```python
df_ipl.dropna(inplace=True)
```

```python
df_ipl.isnull().sum()
```

```python
df_ipl.duplicated().sum()
```

```python
df_ipl.describe()
```

```python
for (colname,colval) in df_ipl.iteritems():
    print('{}:{}'.format(colname,df_ipl[colname].nunique()))
```

```python
for (colname,colval) in df_ipl.iteritems():
    if df_ipl[colname].dtype==object:
        print('{}:{}'.format(colname,df_ipl[colname].unique()))
```

```python
df_ipl['winner']=df_ipl['winner'].replace('Rising Pune Supergiants','Rising Pune Supergiant')
df_ipl['venue']=df_ipl['venue'].replace('M. Chinnaswamy Stadium','M Chinnaswamy Stadium')
```

# Univariate Analysis

### Seasons which had most number of matches

```python
df_ipl['Season'].value_counts()
```

```python
plt.figure(figsize=(17,5))
ax=sns.countplot(df_ipl['Season'],order=['IPL-2008','IPL-2009','IPL-2010','IPL-2011','IPL-2012','IPL-2013','IPL-2014','IPL-2015','IPL-2016','IPL-2017','IPL-2018','IPL-2019'])
plt.title('Number of Matches played per season')
ax.bar_label(ax.containers[0], label_type='edge');
```

**Observation** Most Matches were played in year **2013**.

```python
plt.figure(figsize=(15,6))
sns.countplot(df_ipl['team1'],hue=df_ipl['team2'],edgecolor='k')
plt.xticks(rotation=90)
plt.legend(loc='upper right')
plt.title('Number of matches played by team vs all other team');
```

## Stadium in which most number of matches were played

```python
plt.figure(figsize=(17,5))
ax=sns.countplot(df_ipl['venue'],order=df_ipl['venue'].value_counts().index[0:])
plt.title('Number of Matches played per stadium')
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge');
```

**Observation** Most of the matches were played in **M Chinnaswamy Stadium** and **Eden Gardens**.

## The team that won most matches

```python
plt.figure(figsize=(17,5))
ax=sns.countplot(df_ipl['winner'],order=df_ipl['winner'].value_counts().index[0:])
plt.title('Number of Matches won by each team')
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge');
```

**Observation:** - **Mumbai Indians** wons maximum number of matches including all the seasons

## Top player of the match winners 

```python
df_ipl['player_of_match'].value_counts()[:5]
```

```python
plt.figure(figsize=(17,5))
ax=sns.countplot(df_ipl['player_of_match'],order=df_ipl['player_of_match'].value_counts().index[0:10],palette='inferno')
plt.xticks(rotation=90)
plt.title('Number of times player is player of the match')
ax.bar_label(ax.containers[0], label_type='edge');
```

**Observation** **Chris Gayle** and **AB de Villers** won most of the player of match titles including all seasons.

## Toss Decision

```python
plt.pie(df_ipl['toss_decision'].value_counts(),labels=df_ipl['toss_decision'].unique(),autopct='%1.2f%%',startangle=90)
plt.title('Toss Decision');
```

**Observation** Almost **60%** of the toss decisions are made to field first. Now let us see how this decision varied over time.

```python
plt.figure(figsize=(17,5))
ax=sns.countplot(x='Season', hue='toss_decision', data=df_ipl,order=['IPL-2008','IPL-2009','IPL-2010','IPL-2011','IPL-2012','IPL-2013','IPL-2014','IPL-2015','IPL-2016','IPL-2017','IPL-2018','IPL-2019'])
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.bar_label(ax.containers[1], label_type='edge');
```

**Observation** It seems during the initial years, teams wanted to bat first. Voila.! Look at the **2016 season**, most of the toss decisions are to field first.

```python
# Since there is a very strong trend towards batting second let us see the win percentage of teams batting second.
num_of_wins = (df_ipl.win_by_wickets>0).sum()
num_of_loss = (df_ipl.win_by_wickets==0).sum()
labels = ["Wins", "Loss"]
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins/total)*100, (num_of_loss/total)*100]
plt.pie(sizes, labels=labels,autopct='%1.1f%%',startangle=90)
plt.title("Win percentage batting second");
```

# Bivariate Analysis

## Teams that have won after selecting batting first 

```python
df_ipl[df_ipl['win_by_runs']!=0]['winner'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df_ipl[df_ipl['win_by_runs']!=0]['winner'].value_counts().index[0:5],y=df_ipl[df_ipl['win_by_runs']!=0]['winner'].value_counts()[0:5],palette="viridis")
plt.xticks(rotation=90)
plt.xlabel('Teams',fontsize=12)
plt.ylabel('Winner',fontsize=12)
plt.title('Teams vs won_during_bat_first',fontsize=15);
```

## Teams that have won by chasing runs

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df_ipl[df_ipl['win_by_wickets']!=0]['winner'].value_counts().index[0:5],y=df_ipl[df_ipl['win_by_wickets']!=0]['winner'].value_counts()[0:5],palette="magma")
plt.xticks(rotation=90)
plt.xlabel('Teams',fontsize=12)
plt.ylabel('Winner',fontsize=12)
plt.title('Teams vs won_during_Chasing',fontsize=15);
```

## Top run scorers in IPL

```python
plt.figure(figsize=(17,5))
temp_df = df_score.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
ax=sns.barplot(x=temp_df['batsman'][:10],y=temp_df['batsman_runs'][:10],palette="turbo")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Top run scorers in IPL",fontsize = 15)
ax.set_xlabel('Batsman');
```

**Observation:**  **Suresh Raina** is leading the chart followed closely by Virat Kohli. David Warner is the top scorer among foreign players.

## Players with more number of boundaries in IPL.

```python
plt.figure(figsize=(17,5))
temp_df2 = df_score.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
ax=sns.barplot(x=temp_df2['batsman'][:10],y=temp_df2['batsman_runs'][:10],palette="turbo")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of boundaries in IPL",fontsize = 15)
ax.set_xlabel('Batsman');
```

**Observation:** Gautam Gambhir has most boundaries.

## Players with more number of sixes in IPL.

```python
plt.figure(figsize=(17,5))
temp_df3 = df_score.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
ax=sns.barplot(x=temp_df3['batsman'][:10],y=temp_df3['batsman_runs'][:10],palette="turbo")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Batsman with most number of sixes in IPL",fontsize = 15)
ax.set_xlabel('Batsman');
```

**Observation:** Chris Gayle has most sixes. Suresh raina has 2nd most both boundaries and sixes

## Bowlers who has bowled most number of balls in IPL.

```python
plt.figure(figsize=(17,5))
temp_df4 = df_score.groupby('batsman')['batsman_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
ax=sns.barplot(x=temp_df4['batsman'][:10],y=temp_df4['batsman_runs'][:10],palette="turbo")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Bowlers who has bowled most number of balls in IPL.",fontsize = 15)
ax.set_xlabel('Batsman');
```

**Observation** Gautam Gambhir has bowled most balls in IPL.

## Top Bowlers - Number of balls bowled in IPL

```python
plt.figure(figsize=(17,5))
temp_df5 = df_score.groupby('bowler')['ball'].agg('count').reset_index().sort_values(by='ball', ascending=False).reset_index(drop=True)
ax=sns.barplot(x=temp_df5['bowler'][:10],y=temp_df5['ball'][:10],palette="plasma")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of balls bowled in IPL",fontsize = 15)
ax.set_xlabel('Bowlers');
```

**Observation** Harbhajan Singh is the top bowler of IPL.

## Top Bowlers - Number of dot balls bowled in IPL

```python
plt.figure(figsize=(17,5))
temp_df6 = df_score.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs', ascending=False).reset_index(drop=True)
ax=sns.barplot(x=temp_df6['bowler'][:10],y=temp_df6['total_runs'][:10],palette="plasma")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of dot balls bowled in IPL",fontsize=15)
ax.set_xlabel('Bowlers');
```

**Observation** P kumar has bowled most dot balls in IPL

## The bowlers who has bowled more number of extras in IPL.

```python
plt.figure(figsize=(17,5))
temp_df7 = df_score.groupby('bowler')['extra_runs'].agg(lambda x: (x>0).sum()).reset_index().sort_values(by='extra_runs', ascending=False).reset_index(drop=True)
ax=sns.barplot(x=temp_df7['bowler'][:10],y=temp_df7['extra_runs'][:10],palette="plasma")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Bowlers with more extras in IPL",fontsize=15)
ax.set_xlabel('Bowlers');
```

**Observation** Malinga has bowled number of extras.

## Most common dismissal types in IPL.

```python
plt.figure(figsize=(17,5))
ax=sns.countplot(df_score['dismissal_kind'],palette="Pastel1")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0], label_type='edge')
ax.set_ylabel("Count")
ax.set_title("Common dismissal types in IPL",fontsize=15);
```

**Observation** **Caught** is the most common dismissal type in IPL followed by Bowled. There are very few instances of hit wicket as well. **'Obstructing the field'** is one of the dismissal type as well in IPL.!
